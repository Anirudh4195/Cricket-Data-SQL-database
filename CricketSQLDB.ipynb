### Using Web API

```python
INDmatches = pd.read_csv('india.csv')
```

```python
INDmatches.head()
```

### Creating a list for Match Description

```python
Description_list=[]
```

### Creating a list for Match ID

```python
Matchid_list=[]
```

```python
Matchid_list = INDmatches.MATCH_ID
```

### Using Web API to gather match description data

```python
from espncricinfo.match import Match
for matchId in INDmatches.MATCH_ID:
    try:
        m = Match(matchId)
        print(m.description)
        Description_list.append(m.description)
    except:
        Description_list.append("Error")
        print("error fetching data for match with id: "+str(matchId))
```

### Creating a table for ID and Description

```python
Match_List = pd.DataFrame(Matchid_list)
Match_List['Description'] = Description_list
```

### WEB SCRAPING USING PYTHON

```python
import bs4
```

```python
from urllib.request import urlopen as uReq
```

```python
from bs4 import BeautifulSoup as soup
```

```python
my_url = 'http://stats.espncricinfo.com/ci/engine/stats/index.html?class=3;filter=advanced;orderby=runs;team=6;template=results;type=batting'
```

#### Grabbing the Webpage

```python
uSite = uReq(my_url)
```

```python
page = uSite.read()
```

```python
uSite.close()
```

#### HTML parsing

```python
page_cricketers = soup(page, "html.parser")
```

```python
Boxes = page_cricketers.findAll("tr", {"class":"data1"})
```

```python
Box= Boxes[0]
```

```python
filename = "T20info.csv"
f = open(filename,"w")

headers = "Name, Matches, Runs, Centuries, Half Centuries\n"

f.write(headers)

for Box in Boxes:
    Name = Box.td.a.text
    Matches = Box.findAll('td')[2].text
    Runs = Box.findAll('td')[5].b.text
    Centuries = Box.findAll('td')[10].text
    Half_Centuries = Box.findAll('td')[11].text

    
    print("Name: " + Name)
    print("Matches: " + Matches)
    print("Runs: " + Runs)
    print("Centuries: " + Centuries)
    print("Half_Centuries: " + Half_Centuries)
    
    f.write(Name + "," +Matches + "," +Runs + "," +Centuries + "," +Half_Centuries +"\n")
f.close()
    
    
```
